```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
data = [
    ("Lana" , " employer" , "Ogero"),
    ( "Online Service" ,"Service 1" , "Ogero"),
    ( "Hybrid Service " ,"Service 2" , "Ogero"),
    ( "In Person Service" ,"Service 3" , "Ogero"),
    ( "Hybrid Service " ,"Forwarded" , "Help Desk"),
    ( "In Person Service" ,"Forwarded" , "Help Desk"), 
    ( "Online Service" ,"Forwarded" , "Help Desk"),
    ("Help Desk" , "1" , "Customer Service"),
    ("Help Desk" , "2" , "Technical Support"),
    #("Help Desk" , "3" , "Finance Department"),
    #("Help Desk" , "4" , "Marketing Department"),
    ("Customer 1", "reported_issue", "Late Delivery"),
    ("Late Delivery", "handled_by", "Customer Support"),
    ("Employee 1", "works_in", "Customer Support"),
    #("Customer 2", "reported_issue", "Login Failure"),
    #("Login Failure", "handled_by", "IT Support"),
    ("Employee 2 ", "works_in", "Technical Support"),
    #("Customer3", "reported_issue", "Wrong Data"),
    #("Wrong Data", "handled_by", "Finance Department"),
    #("Employee 3", "works_in", "Finance Department"),
    #("Fiber Service", "promoted_by", "Marketing Department"),
    #("Tech Company", "launched_by", "Marketing Department"),
   # ("Employee 4", "works_in", "Marketing Department"),
]

    
```

```python
Graph = nx.DiGraph()

for subj, rel, obj in data:
    Graph.add_edge(subj, obj, label=rel)
```

```python
plt.figure(figsize=(10, 6))

pos = nx.spring_layout(Graph, seed=42, k =6)
nx.draw(Graph, pos, with_labels=True, node_color="skyblue", node_size=2000, font_size=10, edge_color="gray")
edge_labels = nx.get_edge_attributes(Graph, "label")
nx.draw_networkx_edge_labels(Graph, pos, edge_labels=edge_labels, font_color="red")

plt.title("Mini Medical Knowledge Graph")
plt.show()
```
